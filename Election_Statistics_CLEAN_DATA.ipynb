```python
from pandas import read_csv,DataFrame,options,set_option
from os import listdir
import csv
```

```python
state_summary_files:list[str] = listdir("C:/Users/michael/Documents/Election_Statistics/2024_LIVE_PRESIDENTIAL_RESULTS/State_Summary_Results")
```

```python
swing_states:list[str] = listdir("C:/Users/michael/Documents/Election_Statistics/2024_LIVE_PRESIDENTIAL_RESULTS/State_Details")
```

```python
def unique_elements(elements:list) -> list:
    final_list:list = []
    for element in elements:
        if(not(element in final_list)):
            final_list.append(element)
    return final_list
```

```python
def add_final_detail_files(files:list[str],state:str):
    for file in files[:]:
        if('FINAL' in file):
            pass
        else:
            filename = f"C:/Users/michael/Documents/Election_Statistics/2024_LIVE_PRESIDENTIAL_RESULTS/State_Details/{state}/{file}"

            with open(filename, newline="") as f_input:
                header = next(csv.reader(f_input))

            if(len(header)>7 and 'virginia' in state.lower()):print(header)
            else:
                header = ['Total_Votes','KH_Vote_Count','KH_Vote_Pct','DT_Vote_Count','DT_Vote_Pct','Pct_Reported']
                header = unique_elements(header)
                # display(read_csv(filename,index_col=0,sep=',',skiprows=1,names=header,on_bad_lines='warn',encoding='unicode_escape').fillna(0))
                read_csv(filename,index_col=0,sep=',',skiprows=1,names=header,on_bad_lines='skip',encoding='unicode_escape').fillna(0)\
                    .to_csv(f"C:/Users/michael/Documents/Election_Statistics/2024_LIVE_PRESIDENTIAL_RESULTS/State_Details/{state}/FINAL_{file}",float_format='%.3f')
```

```python
for file in state_summary_files[:]:
    if('FINAL' in file):
        pass
    else:
        filename = f"C:/Users/michael/Documents/Election_Statistics/2024_LIVE_PRESIDENTIAL_RESULTS/State_Summary_Results/{file}"

        with open(filename, newline="") as f_input:
            header = next(csv.reader(f_input))
        header = ['Datetime_Extraction','KH_Vote_Count','KH_Vote_Pct','DT_Vote_Count','DT_Vote_Pct','Pct_Reported']
        header = unique_elements(header)
        # display(read_csv(filename,index_col=0,sep=',',skiprows=1,names=header,on_bad_lines='warn',encoding='unicode_escape').fillna(0))
        read_csv(filename,index_col=0,sep=',',skiprows=1,names=header,on_bad_lines='warn',encoding='unicode_escape').fillna(0)\
            .to_csv(f"C:/Users/michael/Documents/Election_Statistics/2024_LIVE_PRESIDENTIAL_RESULTS/State_Summary_Results/FINAL_{file}",float_format='%.3f')
```

```python
for state in swing_states[:]:
    print(state)
    files:list[str] = listdir(f"C:/Users/michael/Documents/Election_Statistics/2024_LIVE_PRESIDENTIAL_RESULTS/State_Details/{state}")
    add_final_detail_files(files,state)
```
